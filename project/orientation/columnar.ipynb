```python
%%sql 

    Create table if not exists Sensor(
        IdS int not null,
        Longitud float not null,
        latitud float not null,
        PRIMARY KEY(IDS)
    );
    Create table if not exists Lectura(
        IdL int not null,
        Fecha date not null,
        IdS int not null,
        foreign key (IdS) references Sensor(IdS) on delete restrict on update cascade,
        Primary Key(IdL)
    );
    Create table if not exists Lectura_Temperatura(
        IdL int not null,
        SURFACE_TEMPERATURE float not null,
        SOIL_TEMPERATURE float not null,
        foreign key (IdL) references Lectura(IdL) on delete restrict on update cascade,
        Primary Key(IdL)
    );
    Create table if not exists Lectura_Aire(
        IdL int not null,
        AIR_TEMPERATURE float not null,
        WIND float not null,
        foreign key (IdL) references Lectura(IdL) on delete restrict on update cascade,
        Primary Key(IdL)
    );
    Create table if not exists Lectura_Precipitaciones(
        IdL int not null,
        precipitation float not null,
        WETNESS float not null,
        foreign key (IdL) references Lectura(IdL) on delete restrict on update cascade,
        Primary KEY(IdL)
    );
    Create table if not exists Lectura_RadiacionSolar(
        IdL int not null,
        SOLAR_RADIATION float not null,
        foreign key (IdL) references Lectura(IdL) on delete restrict on update cascade,
        Primary Key(IdL)
    );
    Create table if not exists Lectura_Humedad(
        IdL int not null,
        RELATIVE_HUMIDITY float not null,
        SOIL_MOISTURE float not null,
        foreign key (IdL) references Lectura(IdL) on delete restrict on update cascade,
        Primary Key(IdL)
    );

```
